# PCA

- 데이터에서 중요한 관계를 discover하는데 도움
- more informative feature을 만들 수 있다.

- pca : 원래 기능으로 데이터를 설명하는 대신 변형된 축으로 데이터를 설명함
    - 주로, x축을 "size", y축을 "shape" 이라 표현함! 


- 또한, pca는 각 componene에서 variation의 양을 말해주고 있다.
- 아래 사진을 통해 Size component가 shape component보다 더 variation이 있음을 확인할 수 있다. 

- **즉, 데이터의 분산을 최대한 보존하면서 서로 직교하는 축을 찾아, 고차원 공간의 데이터를 저차원 공간으로 변환하는 기법**!

# PCA for feature engineering


- components 는 variation을 알려주고 각 components에 대해 MI score 를 구할수 있기에 어떤 varaition의 종류가 가장 타겟을 잘 예측하는지 알 수 있다.

- 차원 감소, 이상치 감지, 노이즈 감소, Decorrelation 에 활용됨


- pca는 numeric feature에만 작동함!
- 또한 scale에 매우 민감하기에 standardize하는 것이 중요
- 이외에도 이상치를 제거하는 것이 중요

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.feature_selection import mutual_info_regression
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor

# 시각화 세팅
plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)


def make_mi_scores(X, y):
    X = X.copy()
    for colname in X.select_dtypes(["object", "category"]):
        X[colname], _ = X[colname].factorize()
    discrete_features = [pd.api.types.is_integer_dtype(t) for t in X.dtypes]
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features, random_state=0)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores


# 3장에서 다룬 함수
def score_dataset(X, y, model=XGBRegressor()):

    for colname in X.select_dtypes(["category", "object"]):
        X[colname], _ = X[colname].factorize()
    score = cross_val_score(
        model, X, y, cv=5, scoring="neg_mean_squared_log_error",
    )
    score = -1 * score.mean()
    score = np.sqrt(score)
    return score

df = pd.read_csv("data/ames.csv")
```

```python
#연속형 변수들만
features = [
    "GarageArea",
    "YearRemodAdd",
    "TotalBsmtSF",
    "GrLivArea",
]

print("Correlation with SalePrice:\n")

#상관관계 도출
print(df[features].corrwith(df.SalePrice))
```

```python
X = df.copy()
y = X.pop("SalePrice")
X = X.loc[:, features]


def apply_pca(X, standardize=True):
    # pca 하기 전 표준화
    if standardize:
        X = (X - X.mean(axis=0)) / X.std(axis=0)
    pca = PCA()
    X_pca = pca.fit_transform(X)   # PCA 매트릭스매트릭스
    
    #데이터프레임으로 pca 결과 저장
    component_names = [f"PC{i+1}" for i in range(X_pca.shape[1])]
    X_pca = pd.DataFrame(X_pca, columns=component_names)
    
    loadings = pd.DataFrame(
        pca.components_.T,  # transpose the matrix of loadings
        columns=component_names,  # so the columns are the principal components
        index=X.columns,  # and the rows are the original features
    )
    return pca, X_pca, loadings

pca, X_pca, loadings = apply_pca(X)
print(loadings)
```

```python
X=df.copy()
score = score_dataset(X, y)
print(f"Your score: {score:.5f} RMSLE")
```

```python
X = df.copy()
y = X.pop("SalePrice")

X = X.join(X_pca)
score = score_dataset(X, y)
print(f"pca를 새로운 변수로 추가하였을때 score: {score:.5f} RMSLE")
```
