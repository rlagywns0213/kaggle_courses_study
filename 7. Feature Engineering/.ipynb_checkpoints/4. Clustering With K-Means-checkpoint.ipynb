# Clustering

- 비지도 학습 알고리즘 : target이 없으며 "feature discovery" technique라 부른다.
- Clustering은 그룹에서의 데이터 point끼리 얼마나 비슷한지 할당하는 것을 의미한다. 



- 해당 그림은 year built와 sale price 두 변수간의 선형관계를 설명하고 있다.
- 즉, clustering 방법을 사용하지 않았다면 설명하기 힘든 복잡한 관계를 clustering을 통해 선형 관계로 해석 가능하다.

# K-means clustering

- feature engineering 적인 면에서 직관적이고 적용하기 쉬운 알고리즘이다.
- **straight-line distance 즉, 유클리디안 거리를 사용해서 similarity 를 측정한다.**
- 데이터셋에서 각각의 point는 어느 cluster의 centroid와 가까우냐에 따라 할당된다.
- "k"는 얼마나 ㅁ낳은 centorid가 생성되는지를 뜻하고, 분석가가 설정 가능하다.


- `n_clusters` : 
    - 미리 정해둔 센트로이드 수만큼 랜덤하게 initailaize한다.
    - 그 후, 가까운 cluster centorid에 points를 할당시키고 그 points가 할당되면서 거리를 최소화하는 centroid로 중심점이 이동된다.
    - 이러한 작업을 iterates 하게 된다.
    - 종종, 초기 랜덤 위치가 잘못된 클러스터링으로 끝나는 경우가 있기에 여러 번 반복(n_init)하고 각 점과 중심 사이의 총 거리가 가장 작은 클러스터링을 최적의 클러스터링으로 반환한다.
    
- `max_iter` : 센트로이드가 더이상 움직임이 없을 때까지 iterate 하거나 max_iter 만큼 iterate 한다.


- k means 알고리즘은 scale에 민감하기에 feature을 rescale 해주기도 함
- feature가 comparable scale 이 아니라면 rescaling 해주는것이 효과적!!

- 예시

1. Latitude and Longitude of cities in California
    - 위도와 경도에 의해 묘사된 natural distance를 왜곡할 수 없다!
2. Lot Area and Living Area of houses in Ames, Iowa
    - 주택의 공간이 평당 피트가 더 가치 있는 경향이 있으므로, rescale할만하다.
3. Number of Doors and Horsepower of a 1989 model car ??
    - 두 feature은 비교할 수 있는 단위가 아니므로 rescaling 해야함

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor

# 시각화 세팅
plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)

# 3장에서 다룬 함수
def score_dataset(X, y, model=XGBRegressor()):

    for colname in X.select_dtypes(["category", "object"]):
        X[colname], _ = X[colname].factorize()
    score = cross_val_score(
        model, X, y, cv=5, scoring="neg_mean_squared_log_error",
    )
    score = -1 * score.mean()
    score = np.sqrt(score)
    return score

df = pd.read_csv("data/ames.csv")
```

```python
#독립변수 분리
X = df.copy()
y = X.pop("SalePrice")

# 변수 select
features = [
    "LotArea",
    "TotalBsmtSF",
    "FirstFlrSF",
    "SecondFlrSF",
    "GrLivArea",
]

# Standardize
X_scaled = X.loc[:, features]
X_scaled = (X_scaled - X_scaled.mean(axis=0)) / X_scaled.std(axis=0)

kmeans = KMeans(n_clusters=10, n_init=10, random_state=0)
X["Cluster"] = kmeans.fit_predict(X_scaled)
X
```

- `pd.melt(data, id_vars=['col_name1', 'col_name1'])` : col_name1, col_name2 을 사용한 데이터 재구조화
    - id_vars 를 기준으로 value_vars를행으로 보낸다.

```python
# 시각화

Xy = X.copy()
#카테고리 부여
Xy["Cluster"] = Xy.Cluster.astype("category")
Xy["SalePrice"] = y

sns.relplot(
    x="value", y="SalePrice", hue="Cluster", col="variable",
    height=4, aspect=1, facet_kws={'sharex': False}, col_wrap=3,
    data=Xy.melt(
        value_vars=features, id_vars=["SalePrice", "Cluster"],
    ),
);
```

```python
score_dataset(X, y)
```
