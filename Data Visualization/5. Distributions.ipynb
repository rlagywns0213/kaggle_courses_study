# 실습


```python
import pandas as pd

cancer_b_data = pd.read_csv('data/cancer_b.csv', index_col='Id')
cancer_m_data = pd.read_csv('data/cancer_m.csv', index_col='Id')
```

```python
cancer_b_data.head()
```

```python
cancer_m_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.distplot(a=cancer_b_data['Area (mean)'], label="Benign", kde=False)
sns.distplot(a=cancer_m_data['Area (mean)'], label="Malignant", kde=False)
plt.legend()
```

- 결과 : 악성 종양(Malignant tumors)가 Benigh tumors보다 더 큰 범위의 value를 가지고 있음

```python
sns.kdeplot(data=cancer_b_data['Radius (worst)'], shade=True, label="Benign")
sns.kdeplot(data=cancer_m_data['Radius (worst)'], shade=True, label="Malignant")
```

- 악성으로 분류할 확률이 더 높다. 악성 종양에 대한 곡선이 양성 종양에 대한 곡선보다 훨씬 높기 때문에
