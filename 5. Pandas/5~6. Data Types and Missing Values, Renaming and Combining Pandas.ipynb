# 데이터 타입 
### 데이터 타입 살펴보기
- `df['columns'].dtype` : 시리즈에서는 dtype!!
- `df.dtypes` : 데이터프레임에서는 dtypes!!

### 데이터 타입 변경하기

- `df['colums'].astype(str)`


```python
import pandas as pd

df = pd.read_csv('./data/winemag-data-130k-v2.csv', index_col='Unnamed: 0')
df.head(5)
```

```python
df.dtypes
```

```python
# 실수임을 반환
df['price'].dtype
```

```python
df['price'] = df['price'].astype('str') #새롭게 저장해야 dtype변함 (str => obejct로 들어감!!)
df.dtypes
```

```python
df['price'] = df['price'].astype('string') # string 
'''
numpy array에서는 모두 같은 바이트 크기를 가지는데 string은 같은 바이트를 가지지 않아 바이트 크기 때문에 string 대신 object를 사용!!
'''

df.dtypes
```

```python
df['price'] = df['price'].astype('float32') #새롭게 저장해야 dtype변함
df['price'].dtype
```
