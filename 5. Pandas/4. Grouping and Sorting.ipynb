# 이론
## `Groupby()`

- map 함수를 통해 df 또는 series를 한번에 처리할 수 있지만, 
- 보통 groupby를 통해 그룹 처리를 한 후에 map 함수를 적용하는 경우가 대다수

## Groupwise analysis


```python
import pandas as pd
pd.set_option("display.max_rows", 2)
reviews = pd.read_csv("data/winemag-data-130k-v2.csv", index_col=0)
reviews
```

#### `reviews.groupby('column').column.count()`
    - `value_counts` 와 같은 역할을 함

```python
pd.set_option("display.max_rows", 22)
reviews.groupby('points').points.count()
```

```python
reviews
```

```python
# points columns으로 집계화 한 후, 최소 price 도출
reviews.groupby('points').price.min()
```

### `apply()`함수와 결합

- 간편하게 원하는 정보 추출 가능!!

```python
reviews.groupby('winery').apply(lambda df: df.title)
```

```python
reviews[reviews['winery']=='Štoka']['title']
```

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

### `groupby()`의 method `agg()`

- 앞에서는 count(), sum() 의 함수를 달리 썼지만, 
- df를 동시에 다른 함수를 적용하게 해주는 .. 집계 같은 역할!

```python
reviews.groupby(['country']).price.agg([len, min, max])
```

## Multi-indexes

- groupby 에서 여러개의 column으로 할 경우, multi indexes 가 생김!

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

```python
mi
```

```python
# 인덱스를 풀어주는 함수

countries_reviewed.reset_index()
```

## Sorting

- 정렬
- `df.sort_values(by='column_name')` : 데이터프레임을 column에 따라 정렬 
    - `ascending=False` : 내림차순 
    - `ascending=True` : 오름차순 (디폴트값!!)
- `df.sort_index()` : 데이터프레임을 인덱스 순으로 재정렬

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_index()
```

```python
# 2개 컬럼으로 정렬 _ country 다음 , len 순으로 정렬
countries_reviewed.sort_values(by=['country', 'len'])
```

# 실습

```python
print(reviews['taster_twitter_handle'].value_counts())
print(type(reviews['taster_twitter_handle'].value_counts()))
```

```python
print(reviews.groupby('taster_twitter_handle').size())
type(reviews.groupby('taster_twitter_handle').size())
```

- value_counts 와 groupby 후 size 또는 groupby 후 count 모두 series 를 반환!!

#### 가장 좋은 와인은? 가격은 오름차순으로 정렬하고, point는 최대

- 두개의 변수를 sort_values하면 각자 다른 정렬을 할 수 없음
    - 오른, 오른 
    - 내림, 내림 만 가능

```python
reviews.sort_values(['price','points'], ascending=False)
```

```python
# 가격으로 groupby 후, points 최대인 경우
reviews.groupby(['price'])['points'].max().sort_index()
```

#### 와인의 `variety` 에서의 최소 최댓값 도출

```python
reviews.groupby('variety').price.agg([min, max])
```

#### 와인의 variety의 최소 최대 정렬
    -  in descending order based on minimum price, then on maximum price 

```python
variety_about = reviews.groupby('variety').price.agg([min, max])
variety_about.sort_values(by=['min','max'], ascending=False)
```

#### groupby 후 mean , describe() 통해 집계별 분석 용이

```python
reviewer_mean_ratings = reviews.groupby('taster_name').points.mean()
reviewer_mean_ratings
```

```python
reviewer_mean_ratings.describe()
```
